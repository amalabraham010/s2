```python
import cmath

def q(a, b, c):
    d = b**2 - 4*a*c
    root1 = (-b + cmath.sqrt(d)) / (2 * a)
    root2 = (-b - cmath.sqrt(d)) / (2 * a)

    if d > 0:
        print("Two real roots:", root1.real, "and", root2.real)
    elif d == 0:
        print("One repeated root:", root1.real)
    else:
        print("Complex roots:", root1, "and", root2)
q(1, 5, 6)
q(1, 4, 4)
q(1, 2, 5)
```

```python
import matplotlib.pyplot as plt
Rs=50
V = 120
Rl_values = range(1, 500)
power_values = [(V**2 * Rl) / ((Rs + Rl)**2) for Rl in Rl_values]

plt.plot(Rl_values, power_values)
plt.xlabel("Load Resistance Rl (ohms)")
plt.ylabel("Power (W)")
plt.title("Power Supplied to Load Resistance Rl")
plt.grid(True)
plt.show()

max_power = max(power_values)
optimal_Rl = Rl_values[power_values.index(max_power)]
print(f"Maximum Power: {max_power:.2f} W at Rl = {optimal_Rl} ohms")
```

```python

import numpy as np
import matplotlib.pyplot as plt

R = 16000
C = 1e-6
frequencies = np.logspace(0, 5, 500)

omega = 2 * np.pi * frequencies

G_LP = 1 / (1 + 1j * omega * R * C)
amplitude_LP = np.abs(G_LP)
phase_LP = np.angle(G_LP, deg=True)

G_HP = (1j * omega * R * C) / (1 + 1j * omega * R * C)
amplitude_HP = np.abs(G_HP)
phase_HP = np.angle(G_HP, deg=True)

plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
plt.loglog(frequencies, amplitude_LP)
plt.title("Low-pass Filter Amplitude Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True, which="both")
plt.subplot(2, 2, 2)
plt.semilogx(frequencies, phase_LP)
plt.title("Low-pass Filter Phase Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (degrees)")
plt.grid(True, which="both")
plt.subplot(2, 2, 3)
plt.loglog(frequencies, amplitude_HP)
plt.title("High-pass Filter Amplitude Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True, which="both")
plt.subplot(2, 2, 4)
plt.semilogx(frequencies, phase_HP)
plt.title("High-pass Filter Phase Response")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (degrees)")
plt.grid(True, which="both")
plt.tight_layout()
plt.show()
```

```python
15
import numpy as np
import matplotlib.pyplot as plt

def square_wave_fourier(t, n_terms, T=1):
    result = np.zeros_like(t)
    for n in range(n_terms):
        coefficient = 4 / (np.pi * (2 * n + 1))
        result += coefficient * np.sin((2 * n + 1) * np.pi * t / T)
    return result


T = 1
t = np.arange(-1.1, 1.1, 0.01)

n_values = [1, 3, 6, 10, 20]

plt.figure(figsize=(10, 6))
for n in n_values:
    x_t = square_wave_fourier(t, n, T)
    plt.plot(t, x_t, label=f'n = {n}')

plt.title("Fourier Series Approximation of Square Wave")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.grid()
plt.show()
```
